```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
from pandas import DataFrame, read_csv
from sklearn.model_selection import train_test_split

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from keras.utils import to_categorical
df = pd.read_csv('../pima_diabetes.csv',names=['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class'])
print(df.shape)
```

## Data clearning process

```python
df[['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']].replace(0,np.NaN)
print (df.isnull().sum())
print(df.dtypes)
```

```python
print(df.groupby('class').size())
print(df.describe(include=None))
```

## Encode Class Value

```python
array = df.values
X = array[:, 0:8]
Y = array[:, 8]
test_size = 0.33
seed = 10
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = test_size, random_state=seed)
```

```python
encoder_train = LabelEncoder()
encoder_train.fit(Y_train)
encoder_Y_train = encoder_train.transform(Y_train)
onehot_Y_train = np_utils.to_categorical(encoder_Y_train)

encoder_test = LabelEncoder()
encoder_test.fit(Y_test)
encoder_Y_test = encoder_test.transform(Y_test)
onehot_Y_test = np_utils.to_categorical(encoder_Y_test)
```

```python
def baseline():
    model = Sequential()
    model.add(Dense(10, input_dim=8 , activation='relu'))
# input = (Dense(hidden layer nodes, input nodes))
    model.add(Dense(2, activation='softmax'))
# output = (Dense(2 output nodes))
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model

model = baseline()
model.fit(X_train, onehot_Y_train, epochs=200, batch_size=10)
```

## Evaluate Model

```python
evaluator = KerasClassifier(build_fn=baseline ,epochs=200, batch_size = 10)
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(evaluator, X_test, onehot_Y_test, cv= kfold)
print("Accuracy_evaluation: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(evaluator, X_train, onehot_Y_train, cv= kfold)
print("Accuracy_training: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
score = model.evaluate(X_test, onehot_Y_test)
print("\n%s: %.2f%%" % (model.metrics_name[1], score[1]*100))
```
